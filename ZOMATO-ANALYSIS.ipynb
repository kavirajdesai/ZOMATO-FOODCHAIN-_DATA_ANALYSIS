```python
import pandas as pd 
#Pandas is a powerful data manipulation and analysis library. 
#It provides data structures like DataFrames, which are essentially tables with rows and columns, 
#similar to spreadsheets or SQL tables.

import numpy as np
#Purpose: NumPy is a library for numerical operations in Python. It provides support for large, multi-dimensional 
#arrays and matrices, along with mathematical functions to operate on these elements.

import matplotlib.pyplot as plt
#Matplotlib is a popular data visualization library. It allows you to create a wide variety of static, 
#animated, and interactive plots and charts.

import seaborn as sns
# Seaborn is a statistical data visualization library that is built on top of Matplotlib. It provides a high-level interface 
#for drawing attractive and informative statistical graphics.

plt.style.use('dark_background')
```

```python
df= pd.read_csv('../PYTHON ANALYSIS/zomato.csv')
```

```python
df.head()
```

```python
df.shape #shows nummber of rows and columns respectively
```

```python
df.columns #shows name of columns
```

```python
df=df.drop(['url', 'address', 'phone', 'dish_liked', 'menu_item', 'reviews_list'], axis=1) #axis 1 is used for dropping and chaging in the same data frame
df.head()
```

```python
df.info() #showcases the information of columns
```

```python
df.drop_duplicates(inplace=True) #use for removing duplicate entries
df.shape
```

```python
df['rate'].unique() #shows all the unique values in the columns
```

```python
#function for handling rate column
#nan means null value
def handlerate(value):
    if (value=='NEW' or value=='-'):
        return np.nan
    else:
        value=str(value).split('/')
        value=value[0]
        return float(value)

#.apply used for applying the particular function to the column
df['rate']=df['rate'].apply(handlerate)
df.head()
```

```python
df['rate'].fillna(df['rate'].mean(), inplace=True) #fillna means filling null values, mean function for rate, inplace true means change the dat frame
df['rate'].isnull().sum()  #show total number of null values
```

```python
df.info()
```

```python
df.dropna(inplace=True)  #drop nullvalues if any
df.head(100)
```

```python
df.rename(columns={'name': 'custName', 'approx_cost(for two people)': 'cost2plates', 'listed_in(type)':'types'}, inplace=True) 
```

```python
df.head()
```

```python
df=df.drop(['listed_in(city)'],axis=1) #dropping a column
```

```python
df.head()
```

```python
df['cost2plates'].unique()
```

```python
#function for handling values in cost of plates
def handlecomma(value):
    value=str(value)
    if ',' in value:
        value=value.replace(',', '')
        return float(value)
    else:
        return float(value)
    
df['cost2plates']=df['cost2plates'].apply(handlecomma)
df['cost2plates'].unique()
```

```python
df.head()
```

```python
df['rest_type'].value_counts() # .value_counts() function counts the total number of values for the following columns
```

```python
rest_type=df['rest_type'].value_counts(ascending=False)
rest_type
```

```python
restTypeLessThan1000= rest_type[rest_type<1000] #initialising new df for restaurants types less than 1000
restTypeLessThan1000
```

```python
#function for handling restaurant types
def handleRestType(value):
    if value in restTypeLessThan1000:
        return 'others'
    else:
        return value
    
df['rest_type']=df['rest_type'].apply(handleRestType)
df['rest_type'].value_counts()
```

```python
df.head()
```

```python
df['location'].value_counts()
```

```python
location=df['location'].value_counts(ascending=False)
location
```

```python
locationLessThan1000= location[location<1000]
locationLessThan1000
```

```python
locationLessThan1000= location[location<1000]

def handleLocation(value):
    if value in locationLessThan1000:
        return 'others'
    else:
        return value
    
df['location']=df['location'].apply(handleLocation)
df['location'].value_counts()
```

```python
df.head()
```

```python
df['cuisines'].value_counts()
```

```python
cuisines=df['cuisines'].value_counts(ascending=False)
cuisinesLessThan500= cuisines[cuisines<500]

def handleCuisines(value):
    if value in cuisinesLessThan500:
        return 'others'
    else:
        return value

df['cuisines']=df['cuisines'].apply(handleCuisines)
df['cuisines'].value_counts()
```
